```python
from bs4 import BeautifulSoup
import requests

def fetch_html_tables(url):
    "Returns a list of tables in the html of url"
    page = requests.get(url)
    bs=BeautifulSoup(page.content)
    tables=bs.find_all('table')
    return tables

tables=fetch_html_tables('https://no.wikipedia.org/wiki/Norges_fylker')
table_html=tables[1]


print(str(table_html)[:1000])
```

```python
def html_to_table(html):
    "Returns the table defined in html as a list"
    table=[]
    for row in html.find_all('tr'):
        r=[]
        cells=row.find_all('td')

        if len(cells)==0:
            cells=row.find_all('th')
            

        for cell in cells:
            cell=format(cell)
            r.append(cell)
        

        table.append(r)
    return table

def format(cell):
    "Returns a string after converting bs4 object cell to clean text"
    if cell.content is None:
        s=cell.text
    elif len(cell.content)==0:
        return ''
    else:
        s=' '.join([str(c) for c in cell.content])
        
    #here you can add additional characters/strings you want to 
    #remove, change punctuations or format the string in other
    #ways:
    s=s.replace('\xa0','')
    s=s.replace('\n','')
    s=s.replace(',','.')
    
    return s

table=html_to_table(table_html)


print(str(table)[:1000])
```

```python
def save_data(file_name,table):
    "Saves table to file_name"
    f=open(file_name,'w')
    for row in table:
        f.write(';'.join(row)+'\n')
    f.close()
    
save_data('Fylker.csv',table)
```

```python
import pandas as pd
Tabell=pd.read_csv('Fylker.csv', delimiter=';')
Tabell=Tabell.rename(columns={'Folketall[b][5]': 'Folketall'})
Tabell=Tabell.rename(columns={'Areal[6]': 'Areal'})
Tabell=Tabell.rename(columns={'Målform[7]': 'Målform'})
Tabell=Tabell.rename(columns={'Ant. komm.': 'Antall komunner'})
Tabell['Navn'] = Tabell['Navn'].replace(['Oslo Oslo'],'Oslo')
Tabell['Navn'] = Tabell['Navn'].replace(['Rogaland Rogaland'],'Rogaland')
Tabell['Navn'] = Tabell['Navn'].replace(['Møre og Romsdal Møre og Romsdal'],'Møre og Romsdal')
Tabell['Navn'] = Tabell['Navn'].replace(['Nordland Nordland (Nordlánnda)'],'Nordland')
Tabell['Navn'] = Tabell['Navn'].replace(['Viken  Viken'],'Viken')
Tabell['Navn'] = Tabell['Navn'].replace(['Innlandet Innlandet'],'Innlandet')
Tabell['Navn'] = Tabell['Navn'].replace(['Vestfold og Telemark Vestfold og Telemark'],'Vestfold og Telemark')
Tabell['Navn'] = Tabell['Navn'].replace(['Agder Agder'],'Agder')
Tabell['Navn'] = Tabell['Navn'].replace(['Vestland Vestland'],'Vestlandet')
Tabell['Navn'] = Tabell['Navn'].replace(['Trøndelag Trøndelag (Trööndelage)'],'Trøndelag')
Tabell['Navn'] = Tabell['Navn'].replace(['Troms Finnmark  Troms og Finnmark(Romsa ja Finnmárku)(Tromssa ja Finmarkku)'],'Troms og Finnmark')
Tabell
```
